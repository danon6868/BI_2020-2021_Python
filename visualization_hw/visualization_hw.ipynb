## Lineplot

## Fasta length distribution

```python
# Generate random fasta file
alphabet = 'ATGC' * 10000
with open('data/example.fasta', 'w') as f:
    for i in range(1000):
        seq_len = int(random.gauss(100, 5)) + int(random.gauss(500, 10))
        name = f'> seq_number_{i + 1}\n'
        seq = ''.join(random.sample(alphabet, seq_len)) + '\n'
        f.write(name)
        f.write(seq)
```

```python
def len_dist_count(path):
    
    seqs_length = []
    with open(path) as f:
        while True:
            name = f.readline()
            seq = f.readline()
            
            if seq != '':
                seqs_length.append(len(seq))

            if not name:
                break
    
    data = np.unique(seqs_length, return_counts=True)
    plt.plot(data[0], data[1] / sum(data[1]))
    plt.scatter(data[0], data[1] / sum(data[1]))
    plt.xlabel('sequence length')
    plt.ylabel('relative count')
```

```python
len_dist_count('example.fasta')
```

## Favorite graph

```python
dt = 0.01
t = np.arange(0, 30, dt)
noise_1 = np.random.randn(len(t))
noise_2 = np.random.randn(len(t))

s1 = np.sin(2 * np.pi * 20 * t) + noise_1
s2 = np.sin(2 * np.pi * 20 * t) + noise_2


fig, axs = plt.subplots(2, 1)
axs[0].plot(t, s1, t, s2)
axs[0].set_xlim(0, 2)
axs[0].set_xlabel('time')
axs[0].set_ylabel('s1 and s2')
axs[0].grid(True)

cxy, f = axs[1].cohere(s1, s2, 256, 1. / dt)
axs[1].set_ylabel('coherence')


fig.tight_layout()
plt.show()
```
